# Project 1 - Hubble Space Telescope images

- **Hubble Space Telescope**: large telescope orbiting Earth which captures high-quality pictures of space. It produced the famous "Pillars of Creation" picture, displaying a cloud of gas hiding newborn stars. ![Pillars of Creation](pictures/project_1/Hubble_Pillars_of_Creation.jpg)
- **Galaxy**: system made up of stars, gas, dust and dark matter, bound together by gravity. They contain millions to trillions of stars. We live in the Milky Way galaxy. ![Milky Way](pictures/project_1/Hubble_Milky_Way.jpg)


```python
# a library is a collection of code saved somewhere else
import matplotlib.pyplot as plt # helps with drawing graphs
from galaxy_datasets.pytorch import GZHubble 
# Galaxy Zoo - stores images of galaxies caught by the Hubble Space Telescope
```

- **Data**: information that is collected, stored, and used for analysis.

```python
data = GZHubble(root = "../data/GZHubble", download=False)
```

- **Eliptical Galaxies**: Galaxies shaped like squashed circles or ovals. They usually have older stars and less gas.
- **Spiral Galaxies**: Galaxies that look like pinwheels with a bright center and arms that swirl outwards. They often have lots of gas and new stars forming in the arms.
- **Supernova**: Super-bright explosion of a star at the end of its life. It spreads heavy elements (like gold) into space. 

![Supernova](pictures/project_1/supernova.gif)

- **AGN**: Active Galactic Nuclei, super-bright center of some galaxies, powered by a giant black hole absorbing gas - leading to making light and jets of particles. 

![AGN](pictures/project_1/AGN_NASA.jpg)

```python
print(data.catalog) # extract all the information for every picture downloaded
# print displays information
```

- **ra**: right ascension, x-axis in the celestial map (horizontal).
- **dec**: declination, y-axis in the celestial map (vertical).
- **jpg**: Joint Photographic-Experts Group, common digital image file format.

```python
image_data = data.catalog.iloc[1000]
# get the information of the specific integer location (iloc)
# integer: whole number, e.g. 2, 0, -5
print(image_data)
```

```python
image_file_loc = image_data["file_loc"]
# get the file location from the file_loc column
print(f"The file is stored in: {image_file_loc}") 
# f - format
# string (str) - anything within speech marks ("")
# f"{variable}" - formats the value of the variable into the string

image = plt.imread(image_file_loc) #  image read - uses plt to load the image from the computer

plt.imshow(image) # image show - plots the image onto a graph
plt.imsave("Pictures/picture.jpg", image) # image save - save the image into the desired name
plt.show() # displays the graph
```

- **Loops**: let the computer repeat actions.
    - **WHILE** : repeats the code until a certain condition is broken.
    - **FOR** : repeats the code for a certain amount of times.

```python
for i in range(1, 6): # 1,2,3,4,5
    # print(i) # display the numbers
    # range(start, end(not included))
    image_data = data.catalog.iloc[i*1000] # 1000,2000,3000,4000,5000

    image_file_loc = image_data["file_loc"]

    image = plt.imread(image_file_loc)

    plt.figure(figsize=(2,2)) # resize the graph into 2x2 inches
    plt.imshow(image)
    plt.title(f"Image Number: {i*1000}")
    plt.axis("off") # no need for numbers
    plt.show()
```

- **Conditional statements**: logical instructions that let the program make decisions.
    - **IF**: checks for a certain condition and carry out instructions.
        - E.g. If the person is younger than 10, set the cinema ticket price to free.
    - **ELIF**: else if, checks for another certain condition when the one above fails. 
        - E.g. If the person is older than 65, set the cinema ticket price to be 50% off.
    - **ELSE**: instructions to follow when none of the conditions above are satisfied. 
        - E.g. Full price for the rest (11 to 64 year olds).

```python
i = 0 # counter- counts the number of pictures loaded/displayed
for row in range(50000):
    image_data = data.catalog.iloc[row]

    if image_data['has-spiral-arms-hubble_yes'] > 20:
        # get only spiral galaxies
        image_file_loc = image_data["file_loc"]

        image = plt.imread(image_file_loc)

        plt.figure(figsize=(2,2))
        plt.imshow(image)
        plt.title(f"Image number: {row}")
        plt.axis("off")
        plt.show()

        i += 1 # increment  - increase little by little
        if i>3: # limit to the first 4 images
            break # stop the loop
```
